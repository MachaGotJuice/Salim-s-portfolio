```python
import math as math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits import mplot3d
```

```python
# =============================================================================
# Définition des variables
# =============================================================================

Rbob=4.5*(10**-2) #Largeur de la bobine
Lbob=8.5*(10**-2) #Longeur de la bobine
μ0=1.257*(10**(-6)) #Constance magnétique
Nbob=500#Nombre de spire
n1=Nbob/Lbob
Nbob2=1000
n2=Nbob2/Lbob
C=0.01 #Capacité du condensateur
R=8 #Résistance
L=0.08 #Induction de la Bobine
E=30 #Tension aux bornes du GBF       
w0=1/((L*C)**0.5)  #Pulsation propre du circuit RLC série
Q=1/R*((L/C)**0.5) #Facteur de qualité
rmoins=-w0/(2*Q)-w0/(2*Q)*((1-4*Q**2)**0.5)
rplus=-w0/(2*Q)+w0/(2*Q)*((1-4*Q**2)**0.5)
l=0.4 #Distance pour les bornes du tracé
M=6.8 #Moment magnétique 
m=0.051 #Masse de l'aimant
d=0.10 #Distance entre les 2 bobines                                                                      
```

```python
# =============================================================================
# Définition des fonctions utiles
# =============================================================================

def i_régime_critique (t):
    r=w0/(2*Q)
    return (-r**2)*E*C*math.exp(r*t)

def i_régime_apériodique (t):
    rmoins=-w0/(2*Q)-w0/(2*Q)*((1-4*Q**2)**0.5)
    rplus=-w0/(2*Q)+w0/(2*Q)*((1-4*Q**2)**0.5)
    return -(rmoins*rplus*E*C)/(rmoins-rplus)*(math.exp(rplus*t)-math.exp(rmoins*t))

def i_régime_pseudo_périodique (t):
    p=-R/(2*L)
    q=w0*((4-(1/Q)**2)**0.5)
    return -E*C*math.exp(p*t)*math.sin(q*t)*((p**2)/q+q)

def i(temps):
    if Q==1/2:
        return i_régime_critique(temps)
    elif Q<1/(2) :
        return i_régime_apériodique(temps)
    elif Q>1/(2):
        return i_régime_pseudo_périodique(temps)
Force=[]
def forcemag(x,t):
    if x>=(2*L+d):
        Retard.append(t)
        Force.append(-M*μ0*n1*(Rbob**2)*i(t)/2*(1/((Rbob**2+x**2)**1.5)-1/((Rbob**2+(x-Lbob)**2)**1.5)))
        return -μ0*n1*(Rbob**2)*i(t)/2*(1/((Rbob**2+x**2)**1.5)-1/((Rbob**2+(x-Lbob)**2)**1.5))-μ0*n2*(Rbob**2)*i(t-Retard[0])/2*(1/(((x-Lbob-d)**2+Rbob**2)**1.5)-1/(((2*Lbob+d-x)**2+Rbob**2)**1.5))
    else :
        return -μ0*n1*(Rbob**2)*i(t)/2*(1/((Rbob**2+x**2)**1.5)-1/((Rbob**2+(x-Lbob)**2)**1.5))
def frottements(u,t):
    if u==0:
        return 0
    else:
        return (-0.25)*9.8
def équation(X,t):
    x,dxdt=X
    u=dxdt
    dudt=M*forcemag(x,t)/m+frottements(u,t)
    return [dxdt,dudt]

```

```python
# =============================================================================
# Définition des listes
# =============================================================================
Retard=[]
T=np.linspace(0,2,1000000)
x0=0.
v0=0

solution=odeint(équation,[x0,v0],T)

position=solution[:,0]
V=solution[:,1]

i0=0
for j in range(1,len(T)):
    if abs(V[j]) < 1/1000000  and T[j] != 0:
        i0 = j
        break
```

```python
# =============================================================================
# Affichage
# =============================================================================
# Affhichage 1 :
plt.plot(T[:i0],position[:i0])
plt.plot(T[i0:],[position[i0] for _ in range(len(T[i0:]))])
plt.title('Evolution théorique de la position pour w0='+str(w0)+', Q='+str(Q)+', x0='+str(x0)+' et v0='+str(v0),pad=20)
plt.xlabel('Temps (en s)')
plt.ylabel('Position (en m)')
plt.show()

# Affichage 2 :
plt.plot(T[:i0],V[:i0])
plt.plot(T[i0:],[0 for _ in range(len(T[i0:]))])
plt.xlabel('Temps (en s)')
plt.ylabel('Vitesse (en m*s^-1)')
plt.title('Evolution de la vitesse pour w0='+str(w0)+', Q='+str(Q)+', x0='+str(x0)+' et v0='+str(v0),pad=20)
plt.show()
```
